```python
import csv
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
sns.lineplot(data=dataset,x="class",y="who")
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=dataset)
plt.show()
```

```python
sns.histplot(x='class',data=dataset)
plt.show()
```

```python
sns.set(style="darkgrid")
sns.kdeplot(x='fare',data=dataset)
plt.show()
```

```python
sns.boxplot(x='pclass',y='age',data=dataset)
plt.show()
```
